```python
#Question 3
def replace_not_and_poor(sentence):
    not_index = sentence.find('not')
    poor_index = sentence.find('poor')
    if not_index != -1 and poor_index != -1 and poor_index > not_index:
        sentence = sentence[:not_index] + 'good' + sentence[poor_index + len('poor'):] 
    return sentence
text = "The lyrics is not that poor"
result = replace_not_and_poor(text)
print(result)
```

```python
#Question 4 Sort dictionary by value
dictionary = {1:'a',2:'c',3:'b',4:'h',5:'e',6:'y',7:'m'}
sorted_dictionary = dict(sorted(dictionary.items(),key=lambda x : x[1]))
print(sorted_dictionary)
```
